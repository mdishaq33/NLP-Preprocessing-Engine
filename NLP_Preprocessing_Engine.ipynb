# Task 1: Conceptual Understanding

1.What is the difference between "Love" and "love" in NLP?

"Love" and "love" are treated as different words in NLP if case sensitivity is not handled.
"Love" may appear at the beginning of a sentence or as a proper noun, while "love" is a general word.
To avoid confusion, text is usually converted to lowercase so both become the same.

2.What happens if stopwords are not removed?

If stopwords are not removed, the text will contain many unnecessary words like "is", "the", "and", which do not add much meaning.
This increases data size and can reduce model performance because important words get less focus.

3.Provide two real-world scenarios where removing stopwords can be harmful.

Removing stopwords can be harmful when they carry important meaning.
For example, in sentiment analysis, words like "not" can change the meaning of a sentence.
Also, in questions like "What is your name?", removing "what" or "is" may make the sentence unclear.

4.Explain the difference between stemming and lemmatization.

Stemming cuts words to their root form by removing endings, which may not always be correct.
Lemmatization converts words into their proper base form using meaning and context.
For example, "running" becomes "run" in both, but stemming may give incorrect forms like "runn".

```python
#Task 2:Build Advanced Preprocessing Function
```

```python
import re

def preprocess_text(text):

    if not text:
        return [], ""

    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\d+', '', text)
    #text = re.sub(r'(.)\1+', r'\1', text)
    #re.sub(r'(.)\1{2,}', r'\1\1', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()

    clean_tokens = []
    for word in tokens:
        if len(word) > 2 or word in ["no", "not"]:
            clean_tokens.append(word)

    clean_text = " ".join(clean_tokens)

    return clean_tokens, clean_text
```

```python
#print(preprocess_text("1"))
print(preprocess_text("I have 2 dogs!!!"))
```

```python
sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
```

```python
for sentence in sentences:

    tokens, clean_text = preprocess_text(sentence)

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))

    if total_tokens > 0:
        avg_length = sum(len(word) for word in tokens) / total_tokens
    else:
        avg_length = 0

    print("Original Text:", sentence)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", clean_text)

    print("Total Tokens:", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Average Token Length:", round(avg_length, 2))

    print("-" * 50)
    print("\n")
```

Q1: Which sentence had most noise?


“The sentence with the most noise is 'Win $$$ now!!! Limited offer!!!' because it contains symbols, repeated punctuation, and promotional text.”

Q2: Most meaningful tokens?

“The sentence 'I am not happy with this' retained meaningful tokens because it clearly expresses sentiment even after preprocessing.”

```python
from collections import Counter
```

```python
all_tokens = []

for sentence in sentences:
    tokens, _ = preprocess_text(sentence)
    all_tokens.extend(tokens)
```

```python
word_freq = Counter(all_tokens)
```

```python
print("Top 10 Most Frequent Words:")
print(word_freq.most_common(10))
```

```python
print("\nTop 5 Least Frequent Words:")
print(word_freq.most_common()[-5:])
```

```python
def full_pipeline(text_list):

    all_tokens = []
    clean_sent = []

    for text in text_list:
        tokens, clean_text = preprocess_text(text)

        all_tokens.extend(tokens)
        clean_sent.append(clean_text)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sent
    }
```

```python
result = full_pipeline(sentences)

print(result)
```
